# Modelo **(Lotka-Volterra)** <a href="https://emoji.gg/emoji/7867-predator"><img src="https://cdn3.emoji.gg/emojis/7867-predator.png" width="40px" height="40px" alt="Predator"></a>


No in√≠cio do s√©culo XX, Alfred J. Lotka, um cientista americano, prop√¥s pela primeira vez um modelo de intera√ß√µes predador-presa em 1920, inicialmente relacionado a rea√ß√µes qu√≠micas. Ele estendeu esse modelo para intera√ß√µes biol√≥gicas, explorando o crescimento de popula√ß√µes de presas e predadores. Paralelamente, Vito Volterra, um matem√°tico italiano, tamb√©m estava investigando din√¢micas semelhantes ap√≥s observa√ß√µes de capturas de peixes. Em 1926, Volterra publicou um conjunto de equa√ß√µes semelhantes, reconhecendo a contribui√ß√£o de Lotka. Devido a essas contribui√ß√µes independentes, o modelo ficou conhecido como "modelo Lotka-Volterra".

O modelo Lotka-Volterra, tamb√©m conhecido como Equa√ß√µes de Predador-Presa, √© um par de equa√ß√µes diferenciais utilizadas para descrever a din√¢mica de intera√ß√µes entre duas esp√©cies em um ecossistema. Especificamente, o modelo descreve a rela√ß√£o entre uma esp√©cie predadora e uma esp√©cie presa ao longo do tempo.

A din√¢mica dessas popula√ß√µes √© governada pelas taxas de crescimento e intera√ß√µes entre elas. O modelo descreve um ciclo de feedback, no qual o aumento na popula√ß√£o de presas leva a um aumento na popula√ß√£o de predadores, que por sua vez reduz a popula√ß√£o de presas, levando a uma redu√ß√£o na popula√ß√£o de predadores, e assim por diante. Isso resulta em oscila√ß√µes peri√≥dicas nas popula√ß√µes das duas esp√©cies.

Vale ressaltar que o modelo Lotka-Volterra √© uma simplifica√ß√£o da complexidade dos ecossistemas reais e n√£o leva em considera√ß√£o fatores como recursos limitados, outras intera√ß√µes entre esp√©cies, migra√ß√£o, mudan√ßas ambientais e outras nuances da ecologia.

No entanto, ele oferece uma base para entender como as intera√ß√µes entre predadores e presas podem influenciar suas popula√ß√µes ao longo do tempo.

O modelo b√°sico consiste em duas equa√ß√µes diferenciais acopladas, uma para a popula√ß√£o de predadores e outra para a popula√ß√£o de presas:

$\dfrac{dx}{dt} = x * (Œ± - Œ≤*y)$

$\dfrac{dy}{dt} = y * (-Œ¥ + Œ≥*x)$

Onde:

x: popula√ß√£o de presas\
y: popula√ß√£o de predadores\
$Œ±$: taxa de crescimento de presas sem predadores\
$Œ≤$: taxa de decr√©scimo da popula√ß√£o de presas devido a preda√ß√£o\
$Œ¥$: taxa de mortalidade de predadores sem presas\
$Œ≥$: taxa de crescimento de predadores devido a preda√ß√£o

De forma geral, tem-se um modelo com N esp√©cies que interagem entre si:

$\dfrac{dy_i}{dt} = Œ±_i * y_i + \sum_{j=1}^{N} Œ≤_{ij} * y_i * y_j$

Onde $y_i$ √© a popula√ß√£o da i-√©sima esp√©cie e $Œ±_i$ √© a taxa de crescimento da i-√©sima esp√©cie. Al√©m disso, as esp√©cies interagem entre si e essa rela√ß√£o √© descrita atrav√©s do par√¢metro $Œ≤_{ij}$, que √© a taxa em que essa rela√ß√£o afeta a popula√ß√£o da i-√©sima esp√©cie.

Essas equa√ß√µes descrevem como as intera√ß√µes entre as popula√ß√µes de presas e predadores afetam seus respectivos crescimentos. Quando a popula√ß√£o de presas aumenta, mais alimentos est√£o dispon√≠veis para os predadores, o que leva ao aumento da popula√ß√£o de predadores. No entanto, um aumento na popula√ß√£o de predadores leva a uma maior press√£o de preda√ß√£o sobre as presas, o que pode resultar em uma diminui√ß√£o de sua popula√ß√£o. Esse ciclo de intera√ß√µes leva a flutua√ß√µes peri√≥dicas nas popula√ß√µes de ambas as esp√©cies.

### Leitura e tratamento de dados

```python
#popula√ß√µes
data = pd.read_csv('climate-impact-predator-prey-populations/data/Planilha_sem_titulo_-_Pagina1 (1).csv')
#temperatura global
data_temp = pd.read_csv('climate-impact-predator-prey-populations/data/data_temp.csv')

#salvando os valores maximos e minimos para printar no final os valores desnormalizados
min = data.min()
max = data.max()

min_lobo = min['lobo']
max_lobo = max['lobo']
min_alce = min['alce']
max_alce = max['alce']

#normalizando os valores
scaler = MinMaxScaler()
data['lobo'] = scaler.fit_transform(data[['lobo']])
data['alce'] = scaler.fit_transform(data[['alce']])

#selecionando cada s√©rie
lobo = data.loc[:,"lobo"]
alce = data.loc[:,"alce"]
anos = data.loc[:,"ano"]
#agora para temperatura global
temp = data_temp.loc[:, "Anomaly"]
ano_temp = data_temp.loc[:, "Year"]

#trasnformando em listas
lobo = lobo.values.tolist()
alce = alce.values.tolist()
anos = anos.values.tolist()
temp = temp.values.tolist()
ano_temp = ano_temp.values.tolist()
```

### Visualizando nossas informa√ß√µes em um gr√°fico para entender a din√¢mica entre as popula√ß√µes atrav√©s dos anos

```python
#desnormalizando apenas para ajustar a escala do gr√°fico
for i in range (len(anos)):
  alce[i] = (alce[i] * (max_alce - min_alce)) + min_alce
  lobo[i] = (lobo[i] * (max_lobo - min_lobo)) + min_lobo
```

```python
#normalizando de volta
for i in range (len(anos)):
  alce[i] = (alce[i] - min_alce) / (max_alce - min_alce)
  lobo[i] = (lobo[i] - min_lobo) / (max_lobo - min_lobo)
```

## Modelagem das equa√ß√µes

### Lotka-Volterra: modelando as equa√ß√µes

```python
#modelo
def model(variables, t, params):

  #presas
  x = variables[0]
  #predadores
  y = variables[1]

  #parametros
  alpha, beta, delta, gamma = params;

  #equa√ß√µes
  dxdt = alpha * x - beta * x * y
  dydt = delta * x * y - gamma * y

  return([dxdt, dydt])
```

```python
def loss_func(params, ano, alce, lobo):
    y0 = [alce[0], lobo[0]]

    t = np.linspace(ano[0], ano[-1],num=len(ano))

    output = odeint(model, y0, t, args=(params,))

    loss = 0

    for i in range(len(ano)):
        data_alce = alce[i]
        model_alce = output[i, 0]
        data_lobo = lobo[i]
        model_lobo = output[i,1]

        res = (data_alce - model_alce)**2 + (data_lobo - model_lobo)**2
        loss += res

    return loss
```

```python
#definindo os par√¢metros
params0 = np.array([1,1,1,1])
minimo = scipy.optimize.fmin(loss_func, params0, args=(anos, alce, lobo))
print(minimo)

alpha, beta, delta, gamma = minimo
params = np.array([alpha, beta, delta, gamma])
```

```python
#valores iniciais das popula√ß√µes
y0 = [alce[0], lobo[0]] # [preys, predators]

#tempo
t = np.linspace(anos[0], anos[-1],num=len(anos))
```

```python
#resolvendo as ODEs
y = odeint(model, y0, t, args=(params,))
```

```python
for i in range (len(y)):
  y[i,0] = (y[i,0] * (max_alce - min_alce)) + min_alce
  y[i,1] = (y[i,1] * (max_lobo - min_lobo)) + min_lobo
```

```python
#plotando os resultados num gr√°fico com a mesma escala
plt.plot(t, y[:,0], color = "red")
plt.plot(t, y[:,1], color = "blue")
plt.xlabel("tempo")
plt.ylabel("Alces e Lobos")
plt.legend(['Alces', 'Lobos'])
plt.title('Gr√°fico 3')
plt.show()
```

```python
#agora utilizando escalas diferentes para analisar o comportamento das curvas
f, ax1 = plt.subplots()
ax1.set_xlabel("ano")
ax1.set_ylabel("alce", color="red")
ax1.plot(t, y[:,0], '-', color = "red")
ax1.tick_params(axis='y', labelcolor="red")

ax2 = ax1.twinx()

ax2.set_ylabel("lobo", color="blue")
plt.plot(t, y[:,1], '-', color = "blue")
ax2.tick_params(axis='y', labelcolor="blue")

f.tight_layout()
plt.title('Gr√°fico 4')
plt.show()
```
